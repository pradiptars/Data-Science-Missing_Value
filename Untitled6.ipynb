```python
null_var = df.isnull().sum()/df.shape[0] * 100
null_var
```

```python
drop_colmns = null_var[null_var>1].keys()
drop_colmns
```

```python
df2 = df.drop(columns=drop_colmns)
```

```python
sns.heatmap(df2.isnull())
```

```python
df3 = df2.dropna()
```

```python
df['property_name'].value_counts()/df.shape[0] * 100, df3['property_name'].value_counts()/df3.shape[0] * 100
```

```python
pd.concat([df['property_name'].value_counts()/df.shape[0] * 100, df3['property_name'].value_counts()/df3.shape[0] * 100], axis=1, keys = ['property_name_org', 'property_name_clean'])
```

```python
def cat_var_dist(var):
  return
  pd.concat([df[var].value_counts()/df.shape[0] * 100, df3[var].value_counts()/df3.shape[0] * 100], axis=1, keys = [var+'_org', var+'clean'])
```
